# Step 1:


```python
import requests
from apiclient.discovery import build
```

# Step 2:

```python
Api_Key = "API KEY"
SpreadSheetId="SPREAD SHEET ID GOES HERE"
```

```python
sheet = build('sheets', 'v4', developerKey=Api_Key)
```

```python
type(sheet)
```

# Step 3:

```python
resp = sheet.spreadsheets().get(spreadsheetId=SpreadSheetId)
```

```python
data = resp.execute()
```

```python
data
```

# Step 4

#### Accessing Title of your Spread Sheet

```python
data["properties"]["title"]
```

#### Accessing the Sheet Name 

```python
data["sheets"][0]["properties"]["title"]
```

# Passing Argument

```python
resp = sheet.spreadsheets().get(spreadsheetId=SpreadSheetId, includeGridData=True)
```

```python
data = resp.execute()
data
```

```python
# ------------------------------------- ROW INDEX---COL INDEX---
data["sheets"][0]["data"][0]["rowData"][0]["values"][0]["userEnteredValue"]["stringValue"]
```

# Better Method 

```python
range_name = 'A:B'
result = sheet.spreadsheets().values().get(
    spreadsheetId=SpreadSheetId,
    range=range_name).execute()
rows = result.get('values', [])
print("{} rows Retrieved ".format(len(rows)))
print(rows)
```
